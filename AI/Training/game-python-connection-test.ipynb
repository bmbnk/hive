```python
from mlagents_envs.environment import UnityEnvironment
import numpy as np
from mlagents_envs.environment import ActionTuple

UNITY_CONNECTION = False
HIVE_FILE_NAME = "Hive"

env = UnityEnvironment(file_name=None if UNITY_CONNECTION else HIVE_FILE_NAME, 
                       seed=1, 
                       side_channels=[])


env.reset()
```

```python
def print_gameboard(obs):
    arr = np.array(obs).reshape(24, 24)
    arr_to_print = ""

    for i in range(arr.shape[0]):
        for j in range(arr.shape[1]):
            arr_to_print += str(int(arr[i][j])) if arr[i][j] > 0 else "_"
            arr_to_print += " "
        arr_to_print += "\n"
        
    print(arr_to_print)
```

```python
def play_move(move):
    action_tuple = ActionTuple()
    action_tuple.add_discrete(np.array(move))

    env.set_actions(list(env.behavior_specs)[0], action_tuple)
    
    env.step()
    
    steps = env.get_steps(behavior_name)
    reward = steps[1].reward
    game_state = steps[0].obs[0][0]
    
    print(f'Reward = {reward}')
    
    if (reward != 0):
        game_state = steps[1].obs[0][0]
        
    print('Game State: ')
    print_gameboard(game_state)
```

```python
behavior_name = list(env.behavior_specs)[0]
print(f"Name of the behavior : {behavior_name}")
spec = env.behavior_specs[behavior_name]
```

```python
example_game_moves = [
    [[1,15,15,0]],
    [[1,15,14,0]],
    [[2,15,17,0]],
    [[2,15,13,0]],
    [[6,16,16,0]],
    [[3,16,14,0]],
    [[7,14,16,0]],
    [[6,14,14,0]],
    [[2,15,16,1]],
    [[6,14,15,0]],
    [[2,16,15,0]],
    [[2,14,14,0]]
]
```

```python
for move in example_game_moves:
    play_move(move)
```
